```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing Dataset

We begin by importing the dataset from the github link and convert the raw file into a Pandas DataFrame.  
To understand the structure and get an initial sense of the data, we display the first five columns of the dataset for the first few rows.  

1.1


```python
#Importing dataset from github
url = "https://raw.githubusercontent.com/incribo-inc/cybersecurity_attacks/refs/heads/main/cybersecurity_attacks.csv"
df = pd.read_csv(url)
df.head()
```

# 1.2
# Next we identify any null or missing values.
Handling missing value is a crucial step for performing anaysis as this is done to emsure accuracy and reliability. If missing values are found, they are either filled with default values or removed with respect to the requirements.

```python
#Returns the sum of null values in each column
df.isna().sum()
```

```python
print("\nDataset information:")
print(df.info())
```

Here, are the columns that contain null values


*   Malware Indicators
*   Alerts/Warnings
*   Proxy Information
*   Firewall Logs
*   IDS/IPS Alerts

Type -> Object








```python
#Handling the null values
df['Malware Indicators'] = df['Malware Indicators'].fillna('None Detected') #replaces null values with "None Detected"
df['Alerts/Warnings'] = df['Alerts/Warnings'].fillna('No Alert') #replaces null values with "No Alert"
df['Proxy Information'] = df['Proxy Information'].fillna('No Proxy') #replaces null values with "No Proxy"
df['Firewall Logs'] = df['Firewall Logs'].fillna('No Firewall Log') #replaces null values with "No Firewal Log"
df['IDS/IPS Alerts'] = df['IDS/IPS Alerts'].fillna('No IDS/IPS Alerts') #replaces null values with "No IDS/IPS Alerts"
```

```python
df.isna().sum()
```

# Now we can see that the dataset contains no null values

# 1.3

```python
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
# Converting timestamp to datetime objects
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# extracting month from the timestamp
df['month'] = df['Timestamp'].dt.month_name()
df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True)
# Group by 'month' and 'attack_type', and count the occurrences
monthly_attacks = df.groupby(['month', 'Attack Type']).size().reset_index(name='count')

#Visualization
pivot_df = monthly_attacks.pivot(index='month', columns='Attack Type', values='count')
pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')

plt.title('Number of Attacks by Month and Attack Type')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# 1.4

```python
#Packet Length Distribution Analysis by Attack Type
plt.figure(figsize=(12, 7))
sns.kdeplot(data=df, x='Packet Length', hue='Attack Type', common_norm=False, fill=False)
plt.title('Packet Length Distribution by Attack Type')
plt.xlabel('Packet Length')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

** Unique Insights that can be derived from the analysis **

1->Critical log fields showed significant missing data before cleaning, including Malware Indicators, IDS/IPS Alerts, and Firewall Logs, suggesting incomplete integration of detection systems during attacks.

2-> All the attacks were distributed evenly over the year without any one dominating overall volume. This suggests that organizations must maintain a balanced defense strategy to against all three

3->March has recorded the highest amount of attacks across all types DDoS, Intrusion and Malware attacks indicating high surge mid-year. On the other hand, December recorded the least reported incidents

```python
monthly_totals = pivot_df.sum(axis=1)
highest_month = monthly_totals.idxmax()
lowest_month = monthly_totals.idxmin()
highest_count = monthly_totals.max()
lowest_count = monthly_totals.min()

print(f"Month with highest incidents: {highest_month} ({highest_count} total attacks)")
print(f"Month with lowest incidents: {lowest_month} ({lowest_count} total attacks)")
# Create a bar graph of the monthly totals
plt.figure(figsize=(12, 6))
monthly_totals.plot(kind='bar', color='skyblue')
plt.title('Total Incidents by Month')
plt.xlabel('Month')
plt.ylabel('Total Number of Incidents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('monthly_incidents_bar_graph.png')
```

4-> While all the three types of attack had similar packet length, DDoS had the highest average (785) length indicating higher payload requirement for high-volume attacks

```python
# Group by attack type and calculate packet length statistics
packet_stats = df.groupby('Attack Type')['Packet Length'].agg(['count', 'mean', 'median', 'std', 'min', 'max'])
print(packet_stats)
```

```python
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=packet_stats.index, y=packet_stats['mean'])

# used to add the exact values on top of the bar
for i, val in enumerate(packet_stats['mean']):
    ax.text(i, val + 5, f"{val:.0f}", ha='center')

plt.title("Avg Packet Length by Attack Type")
plt.ylabel("Avg Length (bytes)")
plt.xlabel("Attack Type")
plt.show()
```

 5-> High packet size variability was observed in all attack types, with standard deviations over 400 bytes, indicating attackers use mixed packet sizes as a potential evasion technique.

```python
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=packet_stats.index, y=packet_stats['std'])
for i, val in enumerate(packet_stats['std']):
    ax.text(i, val + 5, f"{val:.0f}", ha='center')

plt.title("Packet Length Std Dev by Attack Type")
plt.ylabel("Standard Deviation (bytes)")
plt.xlabel("Attack Type")
plt.show()
```
